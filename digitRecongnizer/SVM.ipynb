```python
import numpy as np
import sklearn
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
source_train_data = pd.read_csv('./data/train.csv')
```

```python
input_size = 28
```

```python
target = source_train_data['label']
source_train_data = source_train_data.drop(['label'], axis=1)
```

```python
def show_number(number_array, target):
    pixel = np.array(number_array)
    pixel = pixel.reshape((28, 28))
    plt.title("Label {}".format(target))
    plt.imshow(pixel,cmap='gray')
    plt.show()
```

```python
for i in range(10):
    show_number(source_train_data.loc[i], target.loc[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
source_train_data = source_train_data / 255.0
```

```python
x_train, x_test , y_train, y_test = train_test_split(source_train_data, target)
```

```python
from sklearn.svm import LinearSVC
from sklearn.linear_model import LinearRegression
```

```python
clf = LinearSVC()
```

```python
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
test_data = pd.read_csv('./data/test.csv')
```

```python
test_data /= 255.0
```

```python
result = clf.predict(test_data)
```

```python
result = pd.DataFrame(result)
```

```python
result.index.name = "ImageId"
```

```python
result.columns = ['Label']
result.index += 1
```

```python
result.to_csv('output.csv')
```
